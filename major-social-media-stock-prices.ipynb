```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.available
plt.style.use('fivethirtyeight')
```

```python
df=pd.read_csv('social_media_stocks_2012-2022.csv')
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
df['Date']=pd.to_datetime(df['Date'])
```

```python
df.set_index('Date',inplace=True)
```

```python
df.head()
```

```python
df.Symbol.value_counts()
```

```python
facebook=df.groupby('Symbol').get_group('FB')
Twitter=df.groupby('Symbol').get_group('TWTR')
insta=df.groupby('Symbol').get_group('ETSY')
SnapChat=df.groupby('Symbol').get_group('SNAP')
pinterest=df.groupby('Symbol').get_group('PINS')
```

```python
facebook[['Close','Volume']].plot(kind='line',alpha=0.6,subplots=True,figsize=(13,10),title='Facebook STocks PRice and The Number Of Shares')
Twitter[['Close','Volume']].plot(kind='line',alpha=0.6,subplots=True,figsize=(13,10),title='Twitter STocks PRice and The Number Of Shares')
insta[['Close','Volume']].plot(kind='line',alpha=0.6,subplots=True,figsize=(13,10),title='Insta STocks PRice and The Number Of Shares')
SnapChat[['Close','Volume']].plot(kind='line',alpha=0.6,subplots=True,figsize=(13,10),title='SnapChat STocks PRice and The Number Of Shares')
pinterest[['Close','Volume']].plot(kind='line',alpha=0.6,subplots=True,figsize=(13,10),title='Pinterest STocks PRice and The Number Of Shares')
plt.show();

```
